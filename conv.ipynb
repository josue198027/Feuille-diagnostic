```python
!pip install tensorflowjs -q

import tensorflow as tf
import tensorflowjs as tfjs
from google.colab import files
import os

print("Sélectionne ton fichier potato_model.keras")
uploaded = files.upload()
nom_modele = list(uploaded.keys())[0]

try:
    print(f"Chargement du modèle : {nom_modele}...")
    # Charger le modèle Keras (v3 ou autre)
    model = tf.keras.models.load_model(nom_modele)

    # Sauvegarder temporairement au format legacy H5
    temp_h5 = "temp_model.h5"
    model.save(temp_h5)
    print(f"Modèle converti temporairement en {temp_h5}")

    # Conversion vers TensorFlow.js
    print("Conversion vers TensorFlow.js en cours...")
    !tensorflowjs_converter --input_format=keras {temp_h5} web_model/

    if os.path.exists('web_model/'):
        print("Succès ! Compression et téléchargement...")
        !zip -r web_model.zip web_model/
        files.download('web_model.zip')
    else:
        print("Erreur : Le dossier 'web_model/' n'a pas été créé.")

finally:
    # Nettoyage
    if os.path.exists("temp_model.h5"):
        os.remove("temp_model.h5")
```
